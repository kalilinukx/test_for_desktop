## Bar plots in matplot library

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
student = {'Bob': 87,"Matt":56,"Sam":27}
student
```

```python
names = list(student.keys())
names
```

```python
values = student.values()
values
```

```python
plt.bar(names,values)
plt.title("Distribution of student marks")
plt.xlabel("Names of student ")
plt.ylabel("Marks of student")
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title("Distribution of student marks")
plt.xlabel("Names of student ")
plt.ylabel("Marks of student")
plt.show()
```

#### Scatter plot in matplot library

```python
x = [1,23,43,55,32,66,86,33]
y = [1,5,7,4,2,6,7,8]
plt.scatter(x,y,marker='*',c='r',s=100)
plt.show()
```

#### Two different plot in the scatter join them together

```python
z= [12,45,87,23,44,56,78,94]
plt.scatter(x,z,marker='*',c='r',s=100)
plt.scatter(z,x,marker='.',c='g',s=300)
plt.show()
```

#### Adding subplots in matplotlib

```python
plt.subplot(1,2,1)
plt.scatter(x,z,marker='*',c='r',s=100)
plt.subplot(1,2,2)
plt.scatter(z,x,marker='.',c='g',s=300)
plt.show()
```
