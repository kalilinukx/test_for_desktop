```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
import matplotlib.pyplot as plt 
```

```python
x = np.arange(1,11)
x
```

```python
y = 3*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("x - label")
plt.ylabel('y -label')
plt.title("Line plot")
# plt.xticks('m')
plt.show()
```

```python
plt.xlabel("x - label")
plt.ylabel('y -label')
plt.title("Line plot")
plt.plot(x,y,color='g',linestyle=':',linewidth=1)
plt.show()
```

```python
plt.plot(x,y,linewidth=3)
plt.show()
```
